```python
ðŸ” Lambda with map(), filter(), sorted()
```

```python
nums = [1,2,4,6]
square= list(map(lambda x:x**2,nums))
print(square)
```

```python
even_nums = list(filter(lambda x: x % 2 == 0, nums))
print(even_nums)  # [2, 4]

# sorted() with lambda
people = [{'name': 'Alice', 'age': 30}, {'name': 'Bob', 'age': 20}]
sorted_people = sorted(people, key=lambda x: x['age'])
print(sorted_people)
```

```python
Creating Arrays
```

```python
# Arrays filled with zeros or ones
zeros = np.zeros((2, 3))
ones = np.ones((3, 2))

print(zeros)
print(ones)
```

```python

# Array with a range
arr = np.arange(0, 10, 2)  # [0, 2, 4, 6, 8]

# Linearly spaced array
lin = np.linspace(0, 1, 5)  # [0. , 0.25, 0.5 , 0.75, 1. ]

# Identity matrix
eye = np.eye(4)
print(eye)
```

```python
Indexing & Slicing
```

```python
a = np.array([10, 20, 30, 40, 50])

# Indexing
print(a[0])        # 10

# Slicing
print(a[1:4])      # [20, 30, 40]

# 2D array indexing
b = np.array([[1, 2, 3], [4, 5, 6]])
print(b[0, 2])     # 3
print(b[:, 1])     # [2, 5]
```

```python
 Array Operations
```

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

# Element-wise operations
print(a + b)       # [5 7 9]
print(a * b)       # [4 10 18]

# Scalar operations
print(a * 2)       # [2 4 6]

# Universal Functions (ufuncs)
np.sqrt(a)
np.sin(a)
np.log(a)
```
